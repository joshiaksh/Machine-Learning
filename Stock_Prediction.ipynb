```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier


import pandas as pd
import numpy as np
#import yfinance as yf

#from google.colab import files
#files.upload()
```

```python
from datetime import date
#Store the data into a variable
df = pd.read_csv('TSLA.csv')

#Set the date as the index
df = df.set_index(pd.DatetimeIndex(df['Date'].values)).head(20)

#Rename the index as date
df.index.name = 'Date'
#Show the data
df
```

```python
#Manipulate the data
#Create the target column
df["DecisionPrice"]= np.where(df['Close'].shift(-1) > df['Close'],1,0)

#Remove the old date column
df = df.drop(columns=['Date'])

#Show the data
df
```

```python
#Split the data set into feature and target
X = df.iloc[:,0:df.shape[1]- 1].values
Y = df.iloc[:,df.shape[1]-1].values
```

```python
#Split the data again into 80% training data and 20% testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
#Create and train the model
tree = DecisionTreeClassifier().fit(X_train,Y_train)

#Show the result of the model
print(tree.score(X_test,Y_test))
```

```python
#Show the predictions
tree_predictions = tree.predict(X_test)
print(tree_predictions)
```
